```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Import data

```python
df = pd.read_csv("../data/final.csv")
```

## Filter most common complaints

```python
# find most abundant complaint type over the first 2 months of 2020
df_2020 = df[df["Created Date"].str.contains("2020")]
# select only first 2 months
df_2020_jan_feb = df_2020[df_2020["Created Date"].str.contains("01/\d\d/2020|02/\d\d/2020")]
# find most abundant complaint type
most_common_complaint_type_jan_feb = df_2020_jan_feb["Complaint Type"].value_counts().head(1)
```

```python
most_common_complaint_type_jan_feb
```

```python
# filter the df_2020 dataset for the most abundant complaint type using mode()
df_2020_jan_feb_most_common = df_2020_jan_feb[df_2020_jan_feb["Complaint Type"].mode()[0] == df_2020_jan_feb["Complaint Type"]]
```

## Plotting

January and Febuary most common complaints

```python
# split the df_2020_most_common dataset into 2 datasets, one for each month
df_2020_most_common_jan = df_2020_jan_feb_most_common[df_2020_jan_feb_most_common["Created Date"].str.contains("01/\d\d/2020")]
df_2020_most_common_feb = df_2020_jan_feb_most_common[df_2020_jan_feb_most_common["Created Date"].str.contains("02/\d\d/2020")]

# create a bar chart for the most common complaint type each month
plt.bar(["Jan", "Feb"], [len(df_2020_most_common_jan), len(df_2020_most_common_feb)])
plt.title("Most Common Complaint Type in Jan and Feb 2020")
plt.xlabel("Month")
plt.ylabel("Number of Complaints")
plt.show()
```

June and July complaints for same complaint type

```python
df_2020_june_july = df_2020[df_2020["Created Date"].str.contains("06/\d\d/2020|07/\d\d/2020")]

df_2020_june_july_most_common = df_2020_june_july[df_2020_june_july["Complaint Type"] == df_2020_jan_feb["Complaint Type"].mode()[0]]

df_2020_most_common_june = df_2020_june_july_most_common[df_2020_june_july_most_common["Created Date"].str.contains("06/\d\d/2020")]
df_2020_most_common_july = df_2020_june_july_most_common[df_2020_june_july_most_common["Created Date"].str.contains("07/\d\d/2020")]

# create a bar chart for the most common complaint type each month
plt.bar(["June", "July"], [len(df_2020_most_common_june), len(df_2020_most_common_july)])
plt.title("Most Common Complaint Type in June and July 2020")
plt.xlabel("Month")
plt.ylabel("Number of Complaints")
plt.show()
```
